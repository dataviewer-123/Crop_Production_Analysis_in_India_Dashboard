```python
# Removing if any whitespaces from columns 
df['Season'] = df['Season'].apply(lambda x: x.strip())
df['Crop'] = df['Crop'].apply(lambda x: x.strip())
df['State_Name'] = df['State_Name'].apply(lambda x: x.strip())
```

```python
#Removing values of Production column which contains 0 and still represents the production of whole district 
#Dropping all rows which contains Production column value as 0

df.drop(df[df['Production']==0].index,inplace=True)
print(f'After removing row which has 0 Production : {df.shape[0]}')
```

```python
# Checking unique crop names, as earlier it was been seen that some of the crop names have their synonyms 

df.Crop.unique()
```

```python
df['Crop'] = df['Crop'].apply(lambda x:x.replace('kapas','Cotton(lint)').replace('Cotton(lint)','Cotton'))
df.replace('Jute & mesta', 'Jute', inplace=True)
df.replace('Mesta','Jute',inplace=True)
df.replace('Mesta','Jute',inplace=True)
```

```python
#Replacing sub-category of pulses with pulses

df['Crop'] = df['Crop'].apply(lambda X:X.replace('Other Kharif pulses', 'pulses')
                             .replace('Other Rabi pulses', 'pulses')
                             .replace('Other misc. pulses', 'pulses')
                             .replace('Pulses total', 'pulses')
                             .replace('Peas & beans (Pulses)', 'pulses')
                             .replace('Moong(Green Gram)', 'pulses')
                             .replace('Urad', 'pulses')
                             .replace('Rajmash Kholar', 'pulses')
                             .replace('Horse-gram', 'pulses')
                             .replace('Khesari', 'pulses')
                             .replace('Masoor', 'pulses')
                             .replace('Lentil', 'pulses')
                             .replace('Ricebean (nagadal)', 'pulses')
                             .replace('Bean', 'pulses')
                             .replace('Arhar/Tur', 'pulses'))



```

```python
#Replacing variety of spices names with other spices

df['Crop'] = df['Crop'].apply(lambda x:x.replace('Black pepper', 'Other Spices')
                             .replace('Cardamom','Other Spices')
                              .replace('Perilla','Other Spices'))
```

```python
# Correcting spelling mistakes in crop names
df['Crop'] = df['Crop'].apply(lambda x:x.replace('Atcanut (Raw)', 'Arecanut')
                              .replace('Arcanut (Processed)','Arecanut'))
                            
```

```python
#Replacing sub-category of fruits into one category as Fruits

df['Crop'] = df['Crop'].apply(lambda x:x.replace('Lemon','Fruits')
                             .replace('Mango','Fruits')
                             .replace('Papaya','Fruits')
                             .replace('Other Fresh Fruits','Fruits')
                             .replace('Orange','Fruits')
                             .replace('Pineapple','Fruits')
                             .replace('Citrus Fruit','Fruits')
                             .replace('Pome Fruit','Fruits')
                             .replace('Grapes','Fruits')
                             .replace('Jack Fruit','Fruits')
                             .replace('Sapota','Fruits')
                             .replace('Pome Granet','Fruits'))
```

```python
df['Crop'] = df['Crop'].apply(lambda x:x.replace('Cashewnut Raw','Cashewnut')
                             .replace('Cashewnut Processed','Cashewnut'))
```

```python
#Replacing sub-category of grams with their main category

df['Crop'] = df['Crop'].apply(lambda x:x.replace('black gram','gram')
                             .replace('Moth','gram')
                             .replace('Blackgram','gram'))
```

```python
#Replacing sub-category of Oilseeds to other oilseeds

df['Crop'] = df['Crop'].apply(lambda x:x.replace('Oilseeds total','other oilseeds')
                             .replace('Niger seed','other oilseeds'))
```

```python
#Replacing sub-category of millets with their main-category

df['Crop'] = df['Crop'].apply(lambda x:x.replace('Other Cereals & Millets','Bajra')
                             .replace('Samai','Bajra')
                             .replace('Small millets','Bajra')
                             .replace('Ragi','Bajra')
                             .replace('Varagu','Bajra'))
```

```python
#Replacing sub-category of vegetables with their main-category

df['Crop'] = df['Crop'].apply(lambda x:x.replace('Redish','Other Vegetables')
                             .replace('Redish','Other Vegetables')
                             .replace('Corainder','Other Vegetables')
                             .replace('Bhindi','Other Vegetables')
                             .replace('Tomato','Other Vegetables')
                             .replace('Cabbage','Other Vegetables')
                             .replace('Carrot','Other Vegetables')
                             .replace('Drum Stick','Other Vegetables')
                             .replace('Cauliflower','Other Vegetables')
                             .replace('Brinjal','Other Vegetables')
                             .replace('Bitter Gourd','Other Vegetables')
                             .replace('Bottle Gourd','Other Vegetables')
                             .replace('Colocosia','Other Vegetables')
                             .replace('Cowpea (Lobia)','Other Vegetables')
                             .replace('pulsess & Mutter (Vegetable)','Other Vegetables'))
```

```python
df['Crop'] = df['Crop'].apply(lambda x:x.replace('Korra','Total foodgrain')
                              .replace('Paddy','Rice')
                              .replace('Niger seed','Sesamum'))
```
